```python
import six
import sys
sys.modules['sklearn.externals.six'] = six
import mlrose
import numpy as np
import matplotlib.pyplot as plt
```

```python
fitness = mlrose.FourPeaks()
state = np.array([1, 1, 1, 0, 1, 0, 0, 1, 0, 0, 0, 0])
fitness.evaluate(state)
```

```python
problem = mlrose.DiscreteOpt(length = 8, fitness_fn = fitness)
```

# Randomized Hill Climbing

```python
best_state,best_fitness,fitness_curve=mlrose.random_hill_climb(problem,curve=True,restarts=20)
print(best_state,best_fitness)
```

```python
plt.plot(fitness_curve)
```

# Simulated Annealing

```python
best_state,best_fitness,fitness_curve=mlrose.simulated_annealing(problem,
                                                                 schedule=mlrose.GeomDecay(init_temp=10.0, decay=0.99, min_temp=0.001),
                                                                 curve=True)
print(best_state,best_fitness)
```

```python
plt.plot(fitness_curve)
```

# Genetic

```python
best_state,best_fitness,fitness_curve=mlrose.genetic_alg(problem,pop_size=4,mutation_prob=0.5,curve=True)
print(best_state,best_fitness)
plt.plot(fitness_curve)
```

# MIMIC

```python
best_state,best_fitness,fitness_curve=mlrose.mimic(problem,pop_size=10,keep_pct=0.5,curve=True)
print(best_state,best_fitness)
plt.plot(fitness_curve)
```
